### Задание  13

##### Продемонстрировать работу с .concat или append

```python
#создание фрейма, являющийся начальным фреймом с разделённым пополам кол-вом индексов первой половины
dataset_part1 = dataset.iloc[0:dataset.shape[0]//2]
dataset_part1
```

```python
#создание фрейма, являющийся начальным фреймом с разделённым пополам кол-вом индексов второй половины
dataset_part2 = dataset.iloc[dataset.shape[0]//2:]
dataset_part2
```

```python
#конкатинация фреймов в один фрейм методом concat()
dataset_concat_edition = pd.concat([dataset_part1, dataset_part2])
dataset_concat_edition
```

```python
#конкатинация фреймов в один фрейм методом append()
dataset_append_edition = dataset_part1.append(dataset_part2)
dataset_append_edition
```

### Задание 14 

##### Проитерировать dataframe построчно .iterrows(). Выполните что-то имеющее смысл внутри цикла

```python
#проитерируем фрейм построчно по определённой стране, с показом информации о дате регистрации и типе вируса
def info_about(country = 'Russia'):
    print(f'Inforamtion about {country}')
    print()
    for index, row in dataset.query(f'location == "{country}"').iterrows():
        print(index, ": Was recorded on ", row['date'], "| Type of variant is ", row['variant'])
```

```python
info_about('Angola')
```
