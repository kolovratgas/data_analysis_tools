## RMSE 

```python
import numpy as np
```

```python
parameters1 = {'iterations': 100, 'depth': 2, 'loss_function': "RMSE", 'verbose': False}
```

```python
cv_dataset = Pool(data=x_tren, label=y_tren)
scores = cv(cv_dataset, parameters1, fold_count=2)
```

```python
mean = np.ones(100)
for i in range(scores.shape[0]):
    dig = np.mean([scores.iloc[i, 1], scores.iloc[i, 2], scores.iloc[i, 3], scores.iloc[i, 4]])
    mean[i] = dig
```

### RMSE: result

```python
pd.DataFrame({'iterations':scores.iloc[0:, 0], 'RMSE_mean':mean}).sort_values('RMSE_mean', ascending=False).head(1)
```

## MAE 

```python
parameters2 = {'iterations': 100, 'depth': 2, 'loss_function': "MAE", 'verbose': False}
cv_dataset2 = Pool(data=x_tren, label=y_tren)
scores2 = cv(cv_dataset2, parameters2, fold_count=2)
```

```python
mean2 = np.ones(100)
for i in range(scores2.shape[0]):
    dig = np.mean([scores2.iloc[i, 1], scores2.iloc[i, 2], scores2.iloc[i, 3], scores2.iloc[i, 4]])
    mean2[i] = dig
```

### MAE: Result 

```python
pd.DataFrame({'iterations':scores2.iloc[0:, 0], 'RMSE_mean':mean}).sort_values('RMSE_mean', ascending=False).head(1)
```
